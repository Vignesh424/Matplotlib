# Creating a Line Chart

```python
from matplotlib import pyplot as plt
import numpy as np
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,6,8,10,12,14,16,18]
plt.plot(x,y, color='g')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Linear Scale")
plt.show()
```

# Create Dashed Style Line Chart

```python
from matplotlib import pyplot as plt
import numpy as np
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,6,8,10,12,14,16,18]
plt.plot(x,y, color='g', linestyle="dashed")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Linear Scale")
plt.show()
```

# Create a Dotted Style Line Chart

```python
from matplotlib import pyplot as plt
import numpy as np
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,6,8,10,12,14,16,18]
plt.plot(x,y, color='', linestyle=":")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Linear Scale")
plt.show()
```

# Create Double Line Chart

```python
rajx=[1,2,3,4,5,6,7,8,9]
rajy=[2,4,6,8,10,12,14,16,18]
siax=[1,2,3,4,5,6,7,8,9]
siay=[3,6,9,12,15,18,21,24,27]
plt.plot(rajx, rajy,color='r', linestyle=":")
plt.plot(siax, siay,color='g',linestyle="--")
plt.title("Student Scale")
plt.show()
```

# Create 3 Styles Line Chart

```python
rajx=[1,2,3,4,5,6,7,8,9]
rajy=[2,4,6,8,10,12,14,16,18]
siax=[1,2,3,4,5,6,7,8,9]
siay=[3,6,9,12,15,18,21,24,27]
alexx=[1,2,3,4,5,6,7,8,9]
alexy=[4,8,12,16,20,24,28,32,36]
plt.plot(rajx, rajy,color='r', linestyle=":")
plt.plot(siax, siay,color='g',linestyle="--")
plt.plot(alexx,alexy,color='b')
plt.title("Student Scale")
plt.show()
```

# Creating a Bar Chart

```python
subjects=["English","Hindi","Marathi","Maths","Science","SS"]
marksofRaj=[45,40,34,39,29,50]
plt.bar(subjects, marksofRaj,color='r')
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Raj's ScoreCard")
plt.show()
```

# Creating a Combined Bar Chart

```python
raj={"English":45, "Hindi":40, "Marathi":32,"Science":30,"SS":49,"Maths":50}
simran={"English":40, "Hindi":45, "Marathi":42,"Science":50,"SS":39,"Maths":30}
plt.bar(raj.keys(),raj.values(), color='r')
plt.bar(simran.keys(),simran.values(), color='b')
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Raj and Simran's Combined ScoreCard")
plt.legend(labels=["Raj","Simran"])
plt.show()
```

# Creating a Multiple Bar Chart

```python
Marks = [[34., 45., 34., 29., 45., 44.],
  [43., 33., 25., 21. ,44., 30.],
  [36., 23., 42., 28., 39., 49.]]
X = np.arange(6)
plt.bar(X + 0.00, Marks[0], color = 'b', width = 0.25)
plt.bar(X + 0.25, Marks[1], color = 'g', width = 0.25)
plt.bar(X + 0.50, Marks[2], color = 'r', width = 0.25)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Raj Tina and Simran's Combined ScoreCard")
plt.legend(labels=["Raj","Tina", "Simran"])
plt.show()
```

# Creating a Horizontal Bar Chart

# Creating ScatterPlot

```python
raj=[23,34,45,40,39,34,29]
subjects=["English","Hindi","Marathi","Science","History","Geography","Maths"]
plt.title("Raj Marks ScatterPlot")
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.scatter(raj,subjects, marker="*", color="g", s=100)
plt.show()
```

# Create a Dual Scatter Plot

```python
x=[1,2,3,4,5,6,7,8]
y=[3,4,5,6,7,8,9,1]
z=[5,2,4,3,6,7,8,9]
plt.title("ScatterPlot")
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(x,y,marker="*", color="g", s=100)
plt.scatter(x,z, marker="+", color="r", s=200)
plt.legend(labels=['XY','XZ'])
plt.show()
```

# Create a Double Scatter Plot

```python
x=[1,2,3,4,5,6,7,8]
y=[3,4,5,6,7,8,9,1]
z=[5,2,4,3,6,7,8,9]
plt.title("ScatterPlot")
plt.xlabel("X")
plt.ylabel("Y")
plt.subplot(1,2,2)
plt.scatter(x,y,marker="*", color="g", s=150)
plt.subplot(1,2,1)
plt.scatter(x,z, marker="+", color="r", s=200)
plt.show()
```

# Create BoxPlot

```python
score1=[20,34,39,45,49]
score2=[34,39,20,34,41]
score3=[34,23,38,19,23]
data=list([score1,score2,score3])
plt.boxplot(data)
plt.show()
```

# Violin Plots

```python
score1=[20,34,39,45,49]
score2=[34,39,20,34,41]
score3=[34,23,38,19,23]
data=list([score1,score2,score3])
plt.violinplot(data)
plt.show()
```

# Create Pie Chart

```python
fruits=['Apple','Mango','Orange','Banana','Chickoo','Grapes']
cost=[24,23,34,31,12,39]
plt.pie(cost, labels=fruits, autopct="%0.f%%", colors=['red','grey','orange','yellow','cyan','green'], radius=1)
plt.show()
```

# Creating Doughnut Chart

```python
fruits=['Apple','Mango','Orange','Banana','Chickoo','Grapes']
cost=[24,23,34,31,12,39]
plt.pie(cost, labels=fruits, autopct="%0.f%%", colors=['red','grey','orange','yellow','cyan','green'], radius=1)
plt.pie([100], colors='w',radius=0.5)
plt.show()
```
